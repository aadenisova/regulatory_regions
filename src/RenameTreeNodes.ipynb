```python
from ete3 import Tree
import pandas as pd
import json

PATH = "/lustre/fs5/vgl/store/adenisova/Inno/upstreams/"

main_tree_file = "cooney_tree.tree" # "shaohong_feng.tree" 

tree_from_file = open(f"{PATH}/{main_tree_file}", "r").read()

# sp_table = pd.read_csv(f"{PATH}/Inno_and_genome.tsv", sep = "\t")
# sp_table["SPECIES_"] = sp_table["SPECIES"].apply(lambda x: '_'.join(x.split()))
# sp_table["NCBI_ID"] = sp_table["cactus"].apply(lambda x: "_".join(x.split("_")[:2]))
# sp_table.head()
```

```python
tree = Tree(tree_from_file, format=1)


all_cactus_species_summary = open(f"{PATH}/all_cactus_species_summary.json", "r")
summary = all_cactus_species_summary.read()

sp_dict = json.loads(summary)

acc_to_sp = dict()
for assembly in sp_dict["assemblies"]:
    organism_name = assembly["assembly"]["biosample"]["description"]["organism"]["organism_name"]
    acc_to_sp['_'.join(organism_name.split()[:2])] = assembly["assembly"]["assembly_accession"]


"""
Се́рая воро́на — вид птиц из рода во́ронов. Иногда чёрную и серую ворону рассматривают как подвиды одного вида — в этом случае чёрную ворону называют Corvus corone corone, а серую — Corvus corone cornix.
"""

acc_to_sp["Corvus_corone"] = acc_to_sp["Corvus_cornix"] 
acc_to_sp.pop("Corvus_cornix")

"""
Linnaeus gave it the binomial name Corvus monedula
"""

acc_to_sp["Corvus_monedula"] = acc_to_sp["Coloeus_monedula"] 
acc_to_sp["Picoides_pubescens"] = acc_to_sp["Dryobates_pubescens"] 

acc_to_sp.pop("Coloeus_monedula")
acc_to_sp.pop("Dryobates_pubescens")

"""
outgroup: 
Chelonia_mydas (Green sea turtle)
Lacerta_agilis (Sand lizard)
"""

acc_to_sp.pop("Chelonia_mydas")
acc_to_sp.pop("Lacerta_agilis")
inv_acc_to_sp = {v: k for k, v in acc_to_sp.items()}

tree.prune(acc_to_sp.keys())

for node in tree:
    node.name = acc_to_sp[node.name]

print(tree)

file_to_save = open(f"{PATH}/shaohong_feng_truncated.tree", "w")
file_to_save.write(tree.write())
file_to_save.close()
    
```

```python
from Bio import Entrez
import time

def get_species_name(ncbi_id):
    Entrez.email = "savouriess2112@gmail.com"  # Замените на ваш email
    try:
        handle = Entrez.esummary(db="assembly", id=ncbi_id)
        record = Entrez.read(handle)
        species_name = record['DocumentSummarySet']['DocumentSummary'][0]['SpeciesName']
        return species_name
    except Exception as e:
        print(f"Error processing {ncbi_id}: {str(e)}")
        return None

ncbi_ids = leafs_cactus

for ncbi_id in ncbi_ids:
    species_name = get_species_name(ncbi_id)
    if species_name:
        print(f"NCBI ID: {ncbi_id}, Species: {species_name}")
    time.sleep(1)  # Добавляем задержку в 1 секунду между запросами
```
